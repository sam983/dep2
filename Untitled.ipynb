```python
df1=df1.drop('mid',axis=1)
df1.head()
```

```python
from datetime import datetime
df1['date']=df1['date'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d'))
```

```python
df1.head()
```

```python
df1=df1.drop('venue',axis=1)
```

```python
df1=pd.get_dummies(df1,drop_first=True)
```

```python
df1.columns
```

```python
df1=df1[['date', 'runs', 'wickets', 'overs', 'runs_last_5', 'wickets_last_5','bat_team_Delhi Daredevils', 'bat_team_Kings XI Punjab',
       'bat_team_Kolkata Knight Riders', 'bat_team_Mumbai Indians',
       'bat_team_Rajasthan Royals', 'bat_team_Royal Challengers Bangalore',
       'bat_team_Sunrisers Hyderabad', 'bowl_team_Delhi Daredevils',
       'bowl_team_Kings XI Punjab', 'bowl_team_Kolkata Knight Riders',
       'bowl_team_Mumbai Indians', 'bowl_team_Rajasthan Royals',
       'bowl_team_Royal Challengers Bangalore',
       'bowl_team_Sunrisers Hyderabad','total']]
```

```python
df1=df1[df1['overs']>5]
```

```python
df1
```

```python
train=df1[df1['date'].dt.year<=2016]
test=df1[df1['date'].dt.year>2016]
```

```python
X_train=train.drop(['total','date'],axis=1)
X_test=test.drop(['total','date'],axis=1)
```

```python
y_train=train['total']
y_test=test['total']
```

```python
import sklearn
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge=Ridge()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40]}
grid=GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
```

```python
grid.fit(X_train,y_train)
```

```python
prediction=grid.predict(X_test)
```
